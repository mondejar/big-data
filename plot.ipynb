## Introducción a Pyplot


``matplotlib.pyplot`` é unha colección de funcions que fan que *matplotlib* funcione como *MATLAB*. Cada *pyplot* realiza algún cambio á figura: ex., crea unha figura, crea algunhas líneas en un gráfico, decora o gráfico co etiquetas, etc.

En ``matplotlib.pyplot`` varios estados preservanse ao largo de chamadas de funcións.

Xerar visualizacións *pyplot* é moi rápido:

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40])
plt.show()
```

```python
x = [1, 5, 8, 10]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.show()
```

## Formateando o estilo de tu gráfico

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],  color='r', marker='o', linestyle='dashed') # Color red, circle marker
plt.axis((0, 6, 0, 20)) # especifica [xmin, xmax, ymin, ymax]
plt.show()
```

## Traballando co multiples figuras e eixes


```python
import numpy as np

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(10, 5))
plt.subplot(121) # nrows, ncols, index
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(122)  # nrows, ncols, index
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

### Subplots

A función subplots crea unha figura e un o varios suplots. Devolve dous obxectos: fig e ax.

Útil para traballar co varios gráficos o personalizalos.

```python
fig, ax = plt.subplots()  # Crea a figura e a gráfica
fig.set_size_inches(10, 6) 

ax.plot(t1, f(t1), 'bo', label='f(t) con t1')  
ax.plot(t2, f(t2), 'k', label='f(t) con t2')   
ax.plot(t2, np.cos(2 * np.pi * t2), 'r--', label='cos(2πt2)') 

# Engadir título e etiquetas
ax.set_title('Gráficas de f(t) y cos(2πt2)')
ax.set_xlabel('Tiempo (t)')
ax.set_ylabel('Amplitud')

# Engadir lenda
ax.legend()

# Gardar a gráfica
fig.savefig("grafica.png")

# Amosar a gráfica
plt.show()
```

# Gráficas co Pandas

Os obxectos ``pd.Series``, ``pd.DataFrame`` ten implementado un método ``plot()`` que permiten representar o seu contido en forma de gráfica. 


```python
import pandas as pd
import numpy as np

longer_ts = pd.Series(np.random.standard_normal(1000), index=pd.date_range("2000-01-01", periods=1000)) # 1000 days
longer_ts.plot()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Ventas": np.random.randint(10, 100, 7),
    "Gastos": np.random.randint(5, 50, 7)},
    index = pd.date_range("2024-01-01", periods=7))

df.plot(y="Gastos")
df.plot(y="Ventas", title = "Ventas", label = "número de Ventas")
df.plot(title = "Gastos e ventas")
df.plot(y="Gastos", kind="pie")
```
